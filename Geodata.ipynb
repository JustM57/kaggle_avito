```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
bad_labels = ['Административное деление Республики Крым', 'Край (Россия)', 'Область (Россия)',
             'Город федерального значения', 'Автономная область (Россия)', 'Автономный округ (Россия)', 'ХМАО']
bad_labels
```

```python
page = requests.get("https://ru.wikipedia.org/wiki/%D0%A1%D1%83%D0%B1%D1%8A%D0%B5%D0%BA%D1%82%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B9%D1%81%D0%BA%D0%BE%D0%B9_%D0%A4%D0%B5%D0%B4%D0%B5%D1%80%D0%B0%D1%86%D0%B8%D0%B8")
soup = BeautifulSoup(page.content, 'html.parser')

a = []
for i in soup.find_all('td')[32:-45]:
    try:
        tmp = i.find('a')['title']
        if tmp.split()[0] != 'Административно-территориальное':
            if tmp not in bad_labels:
                a.append(tmp)
    except:
        try:
            tmp = i['data-sort-value']
            if tmp != '':
                a.append(tmp)
        except:
            pass
region = []
city = []
people = []
for i in range(len(a)//4):
    region.append(a[i*4])
    people.append(a[i*4+2])
    city.append(a[i*4+3])
region = pd.Series(region)
people = pd.Series(people)
city = pd.Series(city)
df = pd.concat([region, people, city], axis = 1)
df
```

```python
page = requests.get("https://ru.wikipedia.org/wiki/%D0%A4%D0%B5%D0%B4%D0%B5%D1%80%D0%B0%D0%BB%D1%8C%D0%BD%D1%8B%D0%B5_%D0%BE%D0%BA%D1%80%D1%83%D0%B3%D0%B0_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B9%D1%81%D0%BA%D0%BE%D0%B9_%D0%A4%D0%B5%D0%B4%D0%B5%D1%80%D0%B0%D1%86%D0%B8%D0%B8")
soup = BeautifulSoup(page.content, 'html.parser')

tmp = []
fo = {}
capital = []
for i, val in enumerate(soup.find_all('div')[10].find_all('a')[60:76]):
    if i % 2 == 0:
        tmp.append(val['title'])
        fo[tmp[-1]] = {}
    else:
        fo[tmp[-1]]['capital'] = val['title']

for i, val in enumerate(soup.find_all('div')[10].find_all('ul')[2:-4]):
    reg = tmp[i]
    for j in val.find_all('a'):
        if j['title'] not in bad_labels:
            try:
                fo[reg]['regions'].append(j['title'])
            except:
                fo[reg]['regions'] = [j['title']]
fo
```

```python
page = requests.get("https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D0%BE%D0%BD%D0%BE%D0%BC%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%B5_%D1%80%D0%B0%D0%B9%D0%BE%D0%BD%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8")
soup = BeautifulSoup(page.content, 'html.parser')

economic = {}
for i in soup.find_all('a')[93:206]:
    if i.get_text() != '' and i.get_text().split()[0] != '[~':
        text = i.get_text()
        if text[-2:] == 'ий' or text[-2:] == 'ый':
            key = text
        else:
            try:
                economic[key].append(text)
            except:
                economic[key] = [text]
economic
```
